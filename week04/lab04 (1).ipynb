```python
import pandas as pd
url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv'

flights = pd.read_csv(url)
```

```python
# 1. When do the first and last flights leave each day?

flights.groupby(['year', 'month', 'day']).dep_time.min()
```

```python
# 1. When do the first and last flights leave each day?

flights.groupby(['year', 'month', 'day']).dep_time.max()
```

```python
# 2. When there is a missing value in dep_time then this is an indication of canceled flight. 
#Find the number of cancelled flights for each (carrier, origin) combination.

def give_count(group):
     return group.isna().sum()

group_carrier_origin = flights.groupby(['carrier', 'origin'])\
                       .agg(\
                           Null_count_by_group = ('dep_time', give_count)
                           )

group_carrier_origin
```
