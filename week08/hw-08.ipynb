The data frame contains notes taken by staff members. Your task is to clean it up to make it usable for analysis. 

### Tasks
* Create start date, end date columns and populate the values (If notes don't include month or day use 1 for missing value) 

i.e: 
    
    - 2018 => 2018-01-01  
    - June 2021 => 2021-06-01  
    - 2020 => 2020-01-01

* Create phone number column (If notes don't include phone numbers, use None)
* Create a column `duration` showing difference between start & end dates in years as a whole number

```python
import re
import pandas as pd
import numpy as np
import datetime
```

```python
# spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_spicy.csv'
less_spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_less_spicy.csv'
```

```python
data = pd.read_csv(less_spicy, index_col=0)
```

```python
data.head(10)
```

```python
data['notes'] = data['notes'].str.lower()
```

```python
data['start_date']=0 
data['end_date']=0
```

```python
startdateregex=re.compile(r'([0-9]{2}(\/|-)[0-9]{2}(\/|-)[0-9]{4})|([0-9]{4}(-|\/)[0-9]{2}(-|\/)[0-9]{2})|([0-9]{4}(\/|-)[0-9]{2})|^[0-9]{4}$|((January|February|March|April|May|June|July|August|September|October|November|December) [0-9]{4})|([0-9]{4}(\/|-)[0-9]{2}(\/|-)[0-9]{2})|([0-9]{2}(\/|-)[0-9]{2}(\/|-)[0-9]{4})|([0-9]{2}(\/)[0-9]{2})|\s[0-9]{4}(\/|-)[0-9]{2}(\/|-)[0-9]{2}|(?: [0-9]{4})|\s[0-9]{4}')
search1= startdateregex.search('notes')
search1
```

```python
data['start_date']=[startdateregex.search(s).group() if startdateregex.search(str(s)) != None else None for s in data['notes'] ]

```

```python
data['start_date']=pd.to_datetime(data['start_date'], errors='coerce')
data
```

```python
phoneNumberRegex = re.compile(r'(((?<=p#)|(?<=pn:)|(?<=p\/n)|(?<=p#)).*?((?=x)|($)))')
data['phone_number']=data['notes'].str.extract(phoneNumberRegex)[0]
data['phone_number'] = data['phone_number'].replace({np.nan: None})
data.head(10)
```

```python
endDateRegex = re.compile(r'((?<=end).*?((?=\/p#)|(?=\/p\/n)|(?=\/P#)\|(?=\/pn)|(?=-p#)|(?=-pn)|(?=-p\/n)|(?=-P#)|(?=p#)|(?=pn)|(?=p\/n)|(?=P#)|($)))')
data['end_date'] = data['notes'].str.extract(endDateRegex)[0]
data['end_date'] = pd.to_datetime(data['end_date'])
```

```python
data.head(20)
```

```python
data['duration'] = pd.DatetimeIndex(data['end_date']).year-pd.DatetimeIndex(data['start_date']).year
data
```
