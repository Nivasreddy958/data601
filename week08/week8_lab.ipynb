__Your Task__

- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

```python
#Find all the answers with 'jupyter' mentioned in it.4430
import re
count = 0
for i, k in enumerate(df.loc[:, 'Q12_Part_4_TEXT']):
    if len(re.findall('[a-zA-Z0-9_.+-]*(j|J)upyter[a-zA-Z0-9_.+-]*', str(k)))!=0:
        print(k)
        count = count +1
print(count)
```
