```python
import pandas as pd
import numpy as np
```

```python
gdp_data=pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)")
gdp_data
```

```python
df2=gdp_data[5]
df2
```

```python
df1=df2[['Country (or dependent territory)','2022','2023','2024','2025','2026','2027']].copy()
df1
```

```python
df1.rename(columns = {'Country (or dependent territory)':'Country'}, inplace = True)
df1
```

```python
df=pd.read_csv(r'/Users/nivasreddy/Desktop/601 HW files/country.csv')
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
df
```

```python
gdp=pd.merge(df,df1, left_on='Country', right_on='Country', how='outer')
gdp
```

```python
gdp['Average_pop']=((gdp['2027']-gdp['2022'])/gdp['2022'])*100
gdp
```

```python
gdp=gdp.replace(np.NaN,None)
gdp
```

```python
gdp=gdp.sort_values(by='Average_pop', ascending = False, ignore_index= True)
gdp
```
